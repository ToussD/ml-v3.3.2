## Fonctions

```python
# TODO : Écrire une fonction `carre(x)` qui retourne x au carré.
# Tester avec carre(4) et carre(-3).

def carre(x):
    return x ** 2

print(carre(4))   # 16
print(carre(-3))  # 9
```

```python
# TODO : Écrire une fonction `est_pair(n)` qui retourne True si n est pair,
# False sinon. Indice : utiliser l'opérateur modulo %.
# Tester avec est_pair(4) et est_pair(7).

def est_pair(n):
    return n % 2 == 0

print(est_pair(4))  # True
print(est_pair(7))  # False
```

```python
# TODO : Écrire une fonction `moyenne(valeurs)` qui retourne la moyenne
# d'une liste de nombres. Que faire si la liste est vide ? Renvoyer None.
# Tester avec [10, 20, 30] et [] (liste vide).

def moyenne(valeurs):
    if len(valeurs) == 0:
        return None
    return sum(valeurs) / len(valeurs)

print(moyenne([10, 20, 30]))  # 20.0
print(moyenne([]))            # None
```

```python
# TODO : Écrire une fonction `saluer(nom, formule="Bonjour")` qui retourne
# la chaîne "<formule>, <nom> !".
# Tester avec saluer("Alice") et saluer("Bob", formule="Salut").

def saluer(nom, formule="Bonjour"):
    return f"{formule}, {nom} !"

print(saluer("Alice"))                # Bonjour, Alice !
print(saluer("Bob", formule="Salut"))  # Salut, Bob !
```

## Compréhensions de liste

```python
# TODO : À partir de nombres = [1, 2, 3, 4, 5], construire en une ligne
# la liste de leurs carrés.

nombres = [1, 2, 3, 4, 5]
carres = [n ** 2 for n in nombres]
print(carres)  # [1, 4, 9, 16, 25]
```

```python
# TODO : À partir de la même liste, construire en une ligne la liste des
# nombres pairs uniquement (filtrage par if dans la compréhension).

pairs = [n for n in nombres if n % 2 == 0]
print(pairs)  # [2, 4]
```

```python
# TODO : À partir de mots = ["chat", "chien", "oiseau", "poisson"],
# construire en une ligne la liste des longueurs de chaque mot.

mots = ["chat", "chien", "oiseau", "poisson"]
longueurs = [len(m) for m in mots]
print(longueurs)  # [4, 5, 6, 7]
```

```python
# TODO : À partir de valeurs = [3, None, 7, None, 12, 5], construire en
# une ligne la liste des valeurs non nulles, multipliées par 10.

valeurs = [3, None, 7, None, 12, 5]
resultat = [v * 10 for v in valeurs if v is not None]
print(resultat)  # [30, 70, 120, 50]
```

## map, filter, lambda

```python
# TODO : À partir de nombres = [1, 2, 3, 4, 5], utiliser map() + lambda
# pour obtenir la liste de leurs cubes (x³).

nombres = [1, 2, 3, 4, 5]
cubes = list(map(lambda x: x ** 3, nombres))
print(cubes)  # [1, 8, 27, 64, 125]
```

```python
# TODO : À partir de la même liste, utiliser filter() + lambda pour
# garder uniquement les nombres strictement supérieurs à 2.

grands = list(filter(lambda x: x > 2, nombres))
print(grands)  # [3, 4, 5]
```

```python
# TODO : Trier la liste mots = ["banane", "kiwi", "pomme", "fraise"]
# par longueur croissante. Indice : sorted(..., key=lambda ...).

mots = ["banane", "kiwi", "pomme", "fraise"]
tries = sorted(mots, key=lambda m: len(m))
print(tries)  # ['kiwi', 'pomme', 'banane', 'fraise']
```
