# Iris: apprentissage semi-supervisé, premiers pas

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.semi_supervised import LabelPropagation
```

## Chargement du jeu de données

```python
iris = load_iris()
```

## Jeux d'apprentissage et de test

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=2
)
```

## Remplacer les labels par -1 en conservant seulement les 30 premiers 

```python
y = y_train.copy()
y[30:] = -1
```

## Utilisation de l'algorithme de propagation des labels

```python
label_prop = LabelPropagation()
label_prop.fit(X_train, y)
```

```python
label_prop.score(X_test, y_test)
```
