# Exercice: Résolution d'une équation du second degré

Implémentez l'algorithme de résolution d'une [équation réelle du second degrè](https://fr.wikipedia.org/wiki/Équation_du_second_degré#Discriminant). 

Vous créerez une fonction `resolve_equation(a, b, c)` prenant pour paramètres ceux de l'équation et retournant le discriminant et les réponses si elles existent.

Vous afficherez la valeur du discriminant et les solutions.
