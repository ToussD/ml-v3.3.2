## Dictionnaires

```python
# TODO : Créer un dictionnaire `personne` avec les clés :
#   - "nom"   : "Martin"
#   - "age"   : 32
#   - "ville" : "Bordeaux"
# Afficher uniquement la valeur de "age".

personne = {"nom": "Martin", "age": 32, "ville": "Bordeaux"}
print(personne["age"])
```

```python
# TODO : Ajouter la clé "email" avec la valeur "martin@exemple.fr".
# Modifier la valeur de "age" pour la passer à 33.

personne["email"] = "martin@exemple.fr"
personne["age"] = 33
print(personne)
```

```python
# TODO : Tenter d'accéder à la clé "telephone" qui n'existe pas, avec
# une valeur par défaut "inconnu" (méthode .get()).

print(personne.get("telephone", "inconnu"))
```

```python
# TODO : Afficher toutes les clés, puis toutes les valeurs du dictionnaire
# (méthodes .keys() et .values()).

print(list(personne.keys()))
print(list(personne.values()))
```

## Conditions

```python
# TODO : Écrire un bloc if/elif/else qui, à partir d'une variable
# `note` (entre 0 et 20), affiche :
#   - "insuffisant"  si note < 10
#   - "passable"     si 10 <= note < 14
#   - "bien"         si 14 <= note < 17
#   - "très bien"    sinon
# Tester avec note = 12.

note = 12
if note < 10:
    print("insuffisant")
elif note < 14:
    print("passable")
elif note < 17:
    print("bien")
else:
    print("très bien")
```

## Boucles

```python
# TODO : Écrire une boucle for qui affiche les nombres de 1 à 10.

for i in range(1, 11):
    print(i)
```

```python
# TODO : Avec une boucle for, calculer la somme des nombres de 1 à 100
# (sans utiliser sum()).

total = 0
for i in range(1, 101):
    total += i
print(total)  # 5050
```

```python
# TODO : Écrire une boucle while qui divise par 2 le nombre n = 100
# tant qu'il est supérieur à 1, et affiche chaque étape.

n = 100
while n > 1:
    n = n / 2
    print(n)
```

```python
# TODO : Parcourir la liste mots = ["chien", "chat", "oiseau", "poisson"]
# et afficher uniquement les mots de plus de 4 lettres.

mots = ["chien", "chat", "oiseau", "poisson"]
for m in mots:
    if len(m) > 4:
        print(m)
```
