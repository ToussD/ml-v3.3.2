### Apply

La fonction [pandas.Series.apply](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.apply.html) est très souvent utilisée. Elle permet d'appliquer une fonction à l'ensemble des valeurs d'une colonne d'un dataframe (qui est de type `Serie`) : Cette fonction peut-être standard (i.e. fournie par Python), ou d'une fonction que vous aurez implémentée selon vos besoin.

* La fonction `apply` retourne une `Serie` (i.e. une colonne) qui est souvent affectée à une nouvelle colonne du dataframe d'origine.
* Elle prend une fonction `thefunc` en paramètre
* `thefunc` sera appelée autant de fois qu'il y a de lignes dans la `Serie`sur laquelle on l'applique
* le paramétre de `thefunc` correspond à la valeur de l'élément en cours de traitement dans la `Serie`en cours


**Exemple 1** :

```python
def nb_mots(s):
    return len(s.split(' '))

dataframe['nb_word'] = dataframe['sentence'].apply(nb_mots)
```

**Exemple 2** :

En déclarant une lambda expression (i.e. une fonction anonyme) : 

```python
dataframe['nb_word'] = dataframe['sentence'].apply(lambda s: len(s.split(' ')))

```
