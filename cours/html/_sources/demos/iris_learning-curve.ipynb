# Iris: courbe d'apprentissage

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import learning_curve
```

## Obtenir les données

```python
iris = load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_data["species"] = iris.target
iris_data = iris_data.sample(frac=0.4)
```

## Courbe d'apprentissage

```python
train_sizes, train_scores, valid_scores = learning_curve(GaussianNB(), iris_data.iloc[:, :4], iris_data["species"])
```

```python
valid_scores
```

```python
train_scores_mean = train_scores.mean(axis=1)
valid_scores_mean = valid_scores.mean(axis=1)
```

```python
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, valid_scores_mean, 'o-', color="g", label="Validation score")
plt.xlabel("Training examples")
plt.ylabel("Score")
plt.legend();
```
