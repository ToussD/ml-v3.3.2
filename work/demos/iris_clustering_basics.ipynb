# Iris: clustering, premiers pas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
iris_data = pd.DataFrame(
    data=iris.data,
    columns=["sepal_length", "sepal_width", "petal_length", "petal_width"]
)
iris_data.head()
```

## Créer un modèle de clustering

```python
kmeans_cluster = KMeans(n_init="auto", n_clusters=3, random_state=1)
kmeans_cluster.fit(iris_data)
```

```python
kmeans_cluster.labels_
```

## Représentation des classes

```python
colormap = np.array(['red','green','blue'])
plt.scatter(
    iris_data.petal_length,
    iris_data.petal_width,
    c=colormap[iris.target],
    s=40
)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title("Classes")
plt.show()
```

## Représentation des clusters

```python
colormap = np.array(['green','red','blue'])
plt.scatter(
    iris_data.petal_length,
    iris_data.petal_width,
    c=colormap[kmeans_cluster.labels_],
    s=40
)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.title("Clusters")
plt.show()
```
