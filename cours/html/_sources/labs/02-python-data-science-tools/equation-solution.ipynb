# Solution: Résolution d'une équation du second degré


Implémentez l'algorithme de résolution d'une [équation réelle du second degrè](https://fr.wikipedia.org/wiki/Équation_du_second_degré#Discriminant). 

Vous créerez une fonction `resolve_equation(a, b, c)` prenant pour paramètres ceux de l'équation et retournant le discriminant et les réponses si elles existent.

Votre programme affichera la valeur du discriminant et les réponses.

```python
import math

def resolve_equation(a, b, c):
    delta = b**2 - 4*a*c
    x1 = (-b + math.sqrt(delta)) / (2*a) if delta >= 0 else None
    x2 = (-b - math.sqrt(delta)) / (2*a) if delta >= 0 else None
    return delta, x1, x2

delta, x1, x2 = resolve_equation(1, 2, 1)
assert delta == 0
assert x1 == -1
assert x1 == x2
```

```python
print(f"Discriminant = {delta}")
if (delta > 0):
    print(f"Deux racines: x1 = {x1}, x1 = {x1}")
elif delta == 0:
    print(f"Racine double: x = {x1}")
else:
    print("Pas de racine")
```
