# Solution: Analyser une courbe d'apprentissage

Nous allons étudier les données `diabètes`

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes

from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from yellowbrick.model_selection import LearningCurve
```

```python
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target
```

```python
print(diabetes.DESCR)
```

## Comparaison d'algorithmes

On évaluera plusieurs algorithmes. On essaiera d'obtenir de "bonnes" valeurs pour les hyperparamètres. Pour cela, on fera varier les hyperparamètres du modèle et on étudiera à chaque fois les courbes d'apprentissage avant et après optimisation

Nous discuterons deux points essentiels:

- Y a-t-il suffisamment de données ?
- Le modèle fait-il de l'overfitting ou de l'underfitting ?

- Algorithmes:
  - [`DecisionTreeRegressor`](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html#sklearn.tree.DecisionTreeRegressor)
  - [`LinearRegression`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression)
  - [`RandomForestRegressor`](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html#sklearn.ensemble.RandomForestRegressor)


### Arbre de décision

```python
model = DecisionTreeRegressor(random_state=2)
visualizer = LearningCurve(model, scoring='r2')
visualizer.fit(X, y)
visualizer.show();
```

```python
model = DecisionTreeRegressor(max_depth=4, max_features=3, random_state=2)
visualizer = LearningCurve(model, scoring='r2')
visualizer.fit(X, y)
visualizer.fit(X, y)
visualizer.show();
```

### Régression linéaire

```python
# TODO : Visualisez les résultats pour LinearRegression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
visualizer = LearningCurve(model, scoring='r2')
visualizer.fit(X, y)
visualizer.show();
```

```python
# TODO : Visualisez les résultats pour un RandomForestRegressor

rdf_reg = RandomForestRegressor(n_jobs=-1)
visualizer = LearningCurve(rdf_reg, scoring='r2')
visualizer.fit(X, y)
visualizer.show();
```

```python
# TODO : Visualisez les résultats pour RandomForestRegressor.
# Attention, la phase d'apprentissage d'un RandomForestRegressor est
# beaucoup plus longue que celle d'un arbre de décision 

rdf_reg = RandomForestRegressor(max_depth=4, max_features=3, min_samples_split=20, n_jobs=-1)
visualizer = LearningCurve(rdf_reg, scoring='r2')
visualizer.fit(X, y)
visualizer.show();
```
