## Tests

```python
# TODO: Ecrire une fonction vérifiant si un tableau contient au moins un élément différent de 0

def not_all_zeros(a):
    return (a != 0).sum() > 0
    # return (a != 0).any()

x = np.array([1, 0, 0, 0])
y = np.array([0, 0, 0, 0])

assert not_all_zeros(x)
assert not not_all_zeros(y)
```

```python
# TODO: Ecrire une fonction vérifiant qu'un tableau ne possède pas de valeurs nulles. On pourra utiliser la fonction np.isnan

def no_null_values(a):
    return not np.isnan(a).any()

x = np.array([1, 2, np.nan, 3])
y = np.array([4, 5, 6, 7])

assert not no_null_values(x)
assert no_null_values(y)
```
