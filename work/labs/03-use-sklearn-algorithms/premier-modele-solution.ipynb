## Charger le dataset

```python
# TODO : Importer load_wine depuis sklearn.datasets et charger le dataset
# dans une variable `wine`.

from sklearn.datasets import load_wine

wine = load_wine()
```

```python
# TODO : Afficher la forme (.shape) de `wine.data` et `wine.target`.
# Combien de lignes ? Combien de features ? Combien de labels ?

print("data   :", wine.data.shape)   # (178, 13)
print("target :", wine.target.shape) # (178,)
# → 178 vins, 13 features, 178 labels
```

```python
# TODO : Afficher `wine.feature_names` et `wine.target_names`.
# Vérifier qu'on a bien 13 features et 3 classes.

print("features :", wine.feature_names)
print("classes  :", wine.target_names)
```

## Préparer X et y

```python
# TODO : Créer les variables `X` (les features) et `y` (la cible).

X = wine.data    # 178 × 13
y = wine.target  # 178

print("X :", X.shape)
print("y :", y.shape)
```

## Découper en train et test

```python
# TODO : train_test_split avec test_size=0.25, random_state=0.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0,
)

print("Train :", X_train.shape, y_train.shape)
print("Test  :", X_test.shape,  y_test.shape)
```

## Baseline avec DummyClassifier

```python
# TODO : DummyClassifier(strategy="most_frequent"), score sur le test.

from sklearn.dummy import DummyClassifier

baseline = DummyClassifier(strategy="most_frequent")
baseline.fit(X_train, y_train)
score = baseline.score(X_test, y_test)
print(f"Baseline : {score:.2%}")
# → autour de 40 % (wine n'est pas parfaitement équilibré)
```

## KNeighborsClassifier

```python
# TODO : KNeighborsClassifier(n_neighbors=1), score sur le test.

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
score = knn.score(X_test, y_test)
print(f"KNN : {score:.2%}")
# → ~75 % : largement au-dessus de la baseline,
#   mais pas exceptionnel (les features wine ne sont pas à la même échelle,
#   ce qui pénalise un KNN basé sur les distances — on en reparlera).
```

## predict et predict_proba

```python
# TODO : Prédire les 5 premiers vins de test, comparer avec y_test[:5].

print("Vraies classes :", y_test[:5])
print("predict        :", knn.predict(X_test[:5]))
```

```python
# TODO : predict_proba sur les mêmes 5 vins.

knn.predict_proba(X_test[:5])
# → avec n_neighbors=1, les probas sont "binaires" :
#   le modèle attribue 1.0 à la classe du seul voisin trouvé.
```

## Bonus — Influence du `random_state`

```python
# TODO : Refaire avec random_state=1.

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=1,
)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print(f"KNN (random_state=1) : {knn.score(X_test, y_test):.2%}")

# Le score change ! Avec un seul split, on a une mesure bruitée.
# C'est pour ça qu'on utilisera plus tard la *cross-validation*,
# qui moyenne plusieurs splits.
```
